```python
# Exercise 3.7: The Mandelbrot Set
from numpy import empty, abs, log
from pylab import imshow, gray, show, xlabel, ylabel, hot, jet

def mandelbrot_iteration(c, max_iter=100):
    """
    Perform Mandelbrot iteration for a given complex number c.
    Returns the number of iterations before |z| > 2, or max_iter if it never exceeds 2.
    """
    z = 0.0 + 0.0j  # Start with z = 0
    
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    
    return max_iter

# Parameters
N = 100  # Grid size - start small for testing
max_iterations = 100
x_min, x_max = -2.0, 2.0
y_min, y_max = -2.0, 2.0

print(f"Computing Mandelbrot set on {N}x{N} grid...")

# Create arrays to store results
mandelbrot_set = empty([N, N], int)

# Calculate spacing
dx = (x_max - x_min) / N
dy = (y_max - y_min) / N

# Iterate over each point in the grid
for i in range(N):
    if i % 10 == 0:
        print(f"Progress: {i}/{N} rows completed")
    
    y = y_min + i * dy
    for j in range(N):
        x = x_min + j * dx
        c = complex(x, y)
        
        # Perform Mandelbrot iteration
        iterations = mandelbrot_iteration(c, max_iterations)
        mandelbrot_set[i, j] = iterations

print("Computation complete!")

# Basic black and white visualization
print("Creating basic black/white Mandelbrot set...")
# Create binary image: black for points in set, white for points outside
binary_set = empty([N, N], float)
for i in range(N):
    for j in range(N):
        if mandelbrot_set[i, j] == max_iterations:
            binary_set[i, j] = 0.0  # Black for points in the set
        else:
            binary_set[i, j] = 1.0  # White for points outside the set

imshow(binary_set, origin="lower", extent=[x_min, x_max, y_min, y_max])
gray()
xlabel("Real axis")
ylabel("Imaginary axis")
show()
```

```python
# Enhanced version: Color by number of iterations
print("Creating colored version based on iteration count...")
imshow(mandelbrot_set, origin="lower", extent=[x_min, x_max, y_min, y_max])
hot()
xlabel("Real axis")
ylabel("Imaginary axis")
show()
```

```python

# High quality version with larger N
print("Computing high-resolution Mandelbrot set...")
N_high = 5000  # Higher resolution
mandelbrot_high = empty([N_high, N_high], int)

dx_high = (x_max - x_min) / N_high
dy_high = (y_max - y_min) / N_high

for i in range(N_high):
    if i % 50 == 0:
        print(f"High-res progress: {i}/{N_high} rows completed")
    
    y = y_min + i * dy_high
    for j in range(N_high):
        x = x_min + j * dx_high
        c = complex(x, y)
        iterations = mandelbrot_iteration(c, max_iterations)
        mandelbrot_high[i, j] = iterations

# High-resolution black and white version
print("Creating high-resolution black/white image...")
binary_high = empty([N_high, N_high], float)
for i in range(N_high):
    for j in range(N_high):
        if mandelbrot_high[i, j] == max_iterations:
            binary_high[i, j] = 0.0
        else:
            binary_high[i, j] = 1.0

imshow(binary_high, origin="lower", extent=[x_min, x_max, y_min, y_max])
gray()
xlabel("Real axis")
ylabel("Imaginary axis")
show()

# Spectacular colored version
print("Creating spectacular colored high-resolution image...")
imshow(mandelbrot_high, origin="lower", extent=[x_min, x_max, y_min, y_max])
jet()
xlabel("Real axis")
ylabel("Imaginary axis")
show()

# Logarithmic coloring for finer detail
print("Creating logarithmic colored version...")
log_mandelbrot = empty([N_high, N_high], float)
for i in range(N_high):
    for j in range(N_high):
        if mandelbrot_high[i, j] == max_iterations:
            log_mandelbrot[i, j] = 0.0
        else:
            log_mandelbrot[i, j] = log(mandelbrot_high[i, j] + 1)

imshow(log_mandelbrot, origin="lower", extent=[x_min, x_max, y_min, y_max])
hot()
xlabel("Real axis")
ylabel("Imaginary axis")
show()

# Zoomed view of interesting region
print("Creating zoomed view of interesting region...")
# Focus on the "seahorse valley" region
x_zoom = [-0.8, -0.4]
y_zoom = [-0.2, 0.2]
N_zoom = 300

mandelbrot_zoom = empty([N_zoom, N_zoom], int)
dx_zoom = (x_zoom[1] - x_zoom[0]) / N_zoom
dy_zoom = (y_zoom[1] - y_zoom[0]) / N_zoom

for i in range(N_zoom):
    if i % 30 == 0:
        print(f"Zoom progress: {i}/{N_zoom} rows completed")
    
    y = y_zoom[0] + i * dy_zoom
    for j in range(N_zoom):
        x = x_zoom[0] + j * dx_zoom
        c = complex(x, y)
        iterations = mandelbrot_iteration(c, 200)  # More iterations for detail
        mandelbrot_zoom[i, j] = iterations

imshow(mandelbrot_zoom, origin="lower", extent=[x_zoom[0], x_zoom[1], y_zoom[0], y_zoom[1]])
jet()
xlabel("Real axis")
ylabel("Imaginary axis")
show()

print("Mandelbrot set visualization complete!")
print("You should see the characteristic 'beetle' or 'bug' shape of the Mandelbrot set.")
```
