```python
# Exercise 3.3: STM surface visualization
from numpy import loadtxt
from pylab import imshow, gray, show, xlabel, ylabel, colorbar

# Read the STM data
print("Reading STM data...")
data = loadtxt("../cpresources/stm.txt", float)

# Basic density plot
print("Creating basic density plot...")
imshow(data)
show()

# Improved density plot with better formatting
print("Creating formatted density plot...")
imshow(data, origin="lower")
gray()
xlabel("x position")
ylabel("y position")
show()

# Enhanced version with colorbar and custom scaling
print("Creating enhanced density plot...")
imshow(data, origin="lower")
gray()
xlabel("Position (arbitrary units)")
ylabel("Position (arbitrary units)")
colorbar()  # Add colorbar to show height scale
show()

# Version with custom extent to show "real" units
# Assuming the surface is measured over a 10nm x 10nm area
print("Creating density plot with physical units...")
surface_size = 10.0  # nanometers
imshow(data, origin="lower", extent=[0, surface_size, 0, surface_size])
gray()
xlabel("x position (nm)")
ylabel("y position (nm)")
colorbar()
show()

# Alternative color schemes for comparison
print("Trying different color schemes...")

# Hot color scheme
imshow(data, origin="lower", extent=[0, surface_size, 0, surface_size])
from pylab import hot
hot()
xlabel("x position (nm)")
ylabel("y position (nm)")
colorbar()
show()

# Final optimized version - clean grayscale with proper labeling
print("Final optimized visualization...")
imshow(data, origin="lower", extent=[0, surface_size, 0, surface_size])
gray()
xlabel("x position (nm)")
ylabel("y position (nm)")
colorbar()
from pylab import title
title("STM Image of Silicon (111) Surface")
show()

# Zoomed-in view of a section of the surface
print("Creating zoomed view of surface section...")
imshow(data, origin="lower", extent=[0, surface_size, 0, surface_size])
gray()
xlabel("x position (nm)")
ylabel("y position (nm)")
colorbar()
title("STM Image of Silicon (111) Surface - Zoomed View")
from pylab import xlim, ylim
xlim(2, 8)  # Focus on middle portion
ylim(2, 8)
show()
```
